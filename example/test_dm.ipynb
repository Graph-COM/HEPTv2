```python
from gnn_tracking_prev.metrics.cluster_metrics import tracking_metrics
```

```python
import numpy as np
import pandas as pd
```

```python
def make_simple_data():
    """
    Create a toy dataset of 6 hits:
      - Cluster 0: all 3 hits from PID 0
      - Cluster 1: 2 hits from PID 1, 1 hit from PID 2
    All pts=1.0, all reconstructable=True
    """
    truth = np.array([0, 0, 0, 1, 1, 2])
    predicted = np.array([0, 0, 0, 1, 1, 1])
    pts = np.ones_like(truth, dtype=float)
    reconstructable = np.ones_like(truth, dtype=bool)
    return truth, predicted, pts, reconstructable
```

```python
truth, predicted, pts, reconstructable = make_simple_data()

res = tracking_metrics(
    truth=truth,
    predicted=predicted,
    pts=pts,
    reconstructable=reconstructable,
    pt_thlds=[0.9],
    predicted_count_thld=3,
)
```
